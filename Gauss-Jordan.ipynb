```python
def print_matrix_aligned(matrix, decimals=2):
    """
    Return
        In ma trận đẹp
    Params
        matrix: ma trận
        decimals: lấy ? số sau dấu phẩy
    """
    formatted_columns = [[] for _ in range(len(matrix[0]))]
    for row in matrix:
        for i, element in enumerate(row[:-1]):
            formatted_columns[i].append("{:>{width}.{decimals}f}".format(element, width=decimals+4, decimals=decimals))
        last_element = row[-1]
        formatted_last_element = "={:>{width}.{decimals}f} ".format(last_element, width=decimals+4, decimals=decimals)
        formatted_columns[-1].append(formatted_last_element)
    max_widths = [max(len(element) for element in column) for column in formatted_columns]
    for row_index in range(len(matrix)):
        for col_index in range(len(matrix[row_index])):
            element = formatted_columns[col_index][row_index]
            print(element.ljust(max_widths[col_index]), end="  ")
        print()
```

```python
def gauss_jordan(mat):
    mat = mat.astype('float32')
    print_matrix_aligned(mat)
    n_rows, n_cols = mat.shape
    row, col = 0, 0
    zero_rows = 0 
    while row < n_rows:
        # Tìm phần tử không bằng 0 đầu tiên trong cột
        while col < n_cols-1 and mat[row][col] == 0:
            col += 1
        # Nếu không tìm thấy, chuyển sang hàng tiếp theo
        if col == n_cols-1:
            row += 1
            continue
        # Đưa phần tử đó thành 1 bằng cách chia hàng đó cho phần tử đó
        print(f'phần tử khác 0 đầu tiên hàng {row} là {mat[row][col]}')
        mat[row] /= mat[row][col]

        # trừ các phần tử khác cùng cột về 0
        for i in range(n_rows):
            if i == row:
                continue
            mat[i] -= mat[i][col] * mat[row]
        print_matrix_aligned(mat)
        print('------------------------------------------------')
        # Tiếp tục với hàng tiếp theo
        row += 1
        col += 1

    # Kiểm tra số chiều của không gian nghiệm
    for i in range(n_cols-1):
        if mat[i][i] == 0 and mat[i][n_cols-1] != 0:
            # Nếu đường chéo chính có phần tử bằng 0 nhưng không phải là hàng zero
            return None
        elif mat[i][i] == 0 and mat[i][n_cols-1] == 0:
            # Nếu đường chéo chính có phần tử bằng 0 và là hàng zero
            zero_rows += 1

    if zero_rows > 0:
        # Trả về số chiều của không gian nghiệm phụ thuộc vào số hàng zero
        return n_cols - 1 - zero_rows
    else:
        # Trả về ma trận kết quả
        return mat[:, -1]
```
